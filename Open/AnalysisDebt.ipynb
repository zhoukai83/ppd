```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt

%matplotlib inline

from IPython.display import display


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 10000)
```

```python
df_listing_info= pd.read_csv("listinginfo.csv", encoding="utf-8")
df_debt_info = pd.read_csv("debtInfo.csv", encoding="utf-8")

df_all = pd.merge(df_listing_info, df_debt_info, on="ListingId")
display(df_all.head(3))
df_all["HighestDebtRatio"] =  (df_all["OwingAmount"] + df_all["Amount"]) / df_all["HighestDebt"]

df_all["HighestPrincipalRatio"] = df_all["Amount"] / df_all["HighestPrincipal"]

df_all["Overdue/NormalCount"] = (df_all["OverdueLessCount"] + df_all["OverdueMoreCount"]) / df_all["NormalCount"]

df_all["SuccessCount/Total"] = df_all["SuccessCount"] / (df_all["WasteCount"] + df_all["CancelCount"] + df_all["FailedCount"] + df_all["SuccessCount"])


display(df_debt_info.head(3))
display(df_listing_info.head(3))
```

```python
display(df_all[df_all["HighestDebtRatio"] > 0.98].shape)

display(df_all["HighestDebtRatio"].describe())

df_all["Days"].describe()
```
