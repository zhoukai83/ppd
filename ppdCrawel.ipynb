```python
def show_ppd_value_counts(df):
    for name in ["resultContent_listing_creditCode",
                "resultContent_listing_months", "resultContent_listing_showRate", "resultContent_loanUse"]:
        display(df[name].value_counts())
        
    print(df["resultContent_listing_creationDate"].min(), df["resultContent_listing_creationDate"].max())

def flat_json(jsonData, parentKey=None):
    flat_dict = {}
    for key in jsonData:
        if type(jsonData[key]).__name__ == 'dict':
            if parentKey:
                temp = flat_json(jsonData[key], parentKey + "_" + key)
                flat_dict = {**flat_dict, **temp}          
            else:
                temp = flat_json(jsonData[key], key)
                flat_dict = {**flat_dict, **temp}
        else:
            if parentKey:
                flat_dict[parentKey + "_" + key] = jsonData[key]
            else:
                flat_dict[key] = jsonData[key]

    return flat_dict
```
