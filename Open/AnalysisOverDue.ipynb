```python
import pandas as pd
from pandas import DataFrame, Series

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 10000)
```

```python
df = pd.read_csv("..\\UI\\UIOverDue.csv", encoding="utf-8")
df
```

```python
df["本次借款后负债"] = df["待还金额"].replace(',','', regex=True).astype(float) + df["借款金额"].replace(',','', regex=True).astype(float)
df["本次借款后负债/历史最高负债"] = df["本次借款后负债"] /  df["历史最高负债"].replace(',','', regex=True).astype(float)

print(df["本次借款后负债/历史最高负债"].describe())
print(df["本次借款后负债/历史最高负债"].value_counts().sort_index())
df
```

```python
temp = df["成功还款次数"].str.extract('(\d+)').astype(int)
print(temp[0].describe())
temp[0].value_counts().sort_index()
```

```python
df["逾期(0-15天)还清次数/成功还款次数"] = df["逾期(0-15天)还清次数"].str.extract('(\d+)').astype(int) / df["成功还款次数"].str.extract('(\d+)').astype(int)
df["逾期(15天以上)还清次数/成功还款次数"] = df["逾期(15天以上)还清次数"].str.extract('(\d+)').astype(int) / df["成功还款次数"].str.extract('(\d+)').astype(int)
df
```

```python
print(df["逾期(15天以上)还清次数/成功还款次数"].describe())
print(df["逾期(15天以上)还清次数/成功还款次数"].value_counts().sort_index())
print("")
print(df["逾期(0-15天)还清次数/成功还款次数"].describe())
df["逾期(0-15天)还清次数/成功还款次数"].value_counts().sort_index()
```

```python
df["借款金额/单笔最高借款金额"] = df["借款金额"].replace(',','', regex=True).astype(float) / df["单笔最高借款金额"].replace(',','', regex=True).astype(float)
print(df["借款金额/单笔最高借款金额"].describe())
print(df["借款金额/单笔最高借款金额"].value_counts().sort_index())
df
```

```python
df_unknow = pd.read_csv("..\\UISimulation\\UISimMain.csv", encoding="utf-8")

def calc_column(df):
    df["本次借款后负债/历史最高负债"] = df["本次借款后负债"] /  df["历史最高负债"]
    df["逾期(0-15天)还清次数/成功还款次数"] = df["逾期（0-15天）还清次数"] / df["成功还款次数"]
    df["逾期(15天以上)还清次数/成功还款次数"] = df["逾期（15天以上）还清次数"] / df["成功还款次数"]
    df["借款金额/单笔最高借款金额"] = df["借款金额"] / df["单笔最高借款金额"]
    return df

df_unknow = calc_column(df_unknow)

print(df_unknow[(df_unknow["逾期(0-15天)还清次数/成功还款次数"] <0.2) & (df_unknow["本次借款后负债/历史最高负债"] <= 1) & (df_unknow["逾期(15天以上)还清次数/成功还款次数"] <0.0001)
  & (df_unknow["借款金额/单笔最高借款金额"] <1) 
  & (df_unknow["网络借贷平台借款余额"] < 8000)
  & (df_unknow["成功还款次数"] > 25)
  ].shape)

print(df_unknow["成功还款次数"].describe())
print(df_unknow["成功还款次数"].value_counts().sort_index())

print(df_unknow["网络借贷平台借款余额"].describe())
print(df_unknow["网络借贷平台借款余额"].value_counts().sort_index())
df_unknow
```
